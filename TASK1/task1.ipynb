```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Load data from the text file
file_path = '/content/drive/MyDrive/task1/Data_1.txt'
data = np.loadtxt(file_path)

# Find peaks
peaks, _ = find_peaks(data)
troughs, _ = find_peaks(-data)

# Plot the data and peaks
plt.plot(data)
plt.plot(peaks, data[peaks], 'x', label='Peaks')
plt.plot(troughs, data[troughs], 'o', label='Troughs')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.show()
```
